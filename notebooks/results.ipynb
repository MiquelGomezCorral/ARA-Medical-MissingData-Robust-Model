```python
rows = []
for model_id, cfg in MODELS.items():
    for var_key, dir_name in cfg.items():
        if var_key not in ("V1", "V2"):
            continue
        path = MODELS_DIR / dir_name / "results.json"
        with open(path) as f:
            data = json.load(f)
        rows.append({
            "model_id": model_id,
            "dataset": cfg["dataset"],
            "mask": cfg["mask"],
            "radiomics": cfg["radiomics"],
            "variant": var_key,
            "f1": data["f1"],
        })

df = pd.DataFrame(rows)
df["label"] = df["mask"] + ("+Rad" if df["radiomics"].iloc[0] else "")
df["radiomics_str"] = df["radiomics"].map({True: "+Rad", False: "-Rad"})
df["mask_label"] = df["mask"] + " " + df["radiomics_str"]

print(f"Loaded {len(df)} variants across {df['model_id'].nunique()} models")
df.head(12)
```
